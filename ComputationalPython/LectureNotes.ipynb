## A timing decorator

The idea is to measure execution time of a given function

```python
import time
import math
```

```python
def slow(x):
    time.sleep(1)
    return math.sqrt(x)
```

```python
slow(2)
```

```python
# doing it by hand

t1 = time.time()
slow(2)
t2 = time.time()
print(f'time spent in slow = {t2-t1}')
```

If we use this code in many places, we can define a decorator for this operation

```python
def identity(f):  # take a function as an input
    return f
# returning the same function: identity decorator
```

```python
identity(slow)(2)
# logic behind syntax:
identity_slow = identity(slow)  # passing the function as argument
identity_slow(2)  # passing the arguments of the function as the argument of this new function
```

```python
def time_me(f):
    # new function definition inside:
    def inner(*args, **kwargs):  # cause we don't know anything about f
        t1 = time.time()
        value = f(*args, **kwargs)  #save its return value
        t2 = time.time()
        print(f'time spent in {f.__name__} = {t2-t1}')
        return value
    return inner  # returning the inner function as the decorator
```

```python
time_me(slow)(2)
```

We have a special syntax for decorators: @decorator above the function to wrap it

```python
# Equivalent to slow = time_me(slow)

@time_me
def slow(x):
    time.sleep(1)
    return math.sqrt(x)
```

```python
slow(2)
```

## debugging decorator

The idea is to track the value the goes into a certain function and the result of that function

```python
def debug(f):
    def inner(*args, **kwargs):
        print(f'Input argumnents to {f.__name__}:', args, kwargs)
        value = f(*args, **kwargs)
        print(f'Return value from {f.__name__}: {value}')
        return value
    return inner
```

```python
@debug
def slow(x):
    time.sleep(1)
    return math.sqrt(x)
```

```python
slow(2)
```

## Combining decorators

```python
@time_me
@debug
def slow(x):
    time.sleep(1)
    return math.sqrt(x)
```

```python
slow(2) # we get some strange behaviour as the funciton name does not propagate to the second decorator
```

```python
# fix it:

from functools import wraps  # to get the right funtion names

def time_me(f):
    @wraps(f)
    def inner(*args, **kwargs):
        t1 = time.time()
        value = f(*args, **kwargs)
        t2 = time.time()
        print(f'time spent in {f.__name__} = {t2-t1}')
        return value
    return inner

def debug(f):
    @wraps(f)
    def inner(*args, **kwargs):
        print(f'Input argumnents to {f.__name__}:', args, kwargs)
        value = f(*args, **kwargs)
        print(f'Return value from {f.__name__}: {value}')
        return value
    return inner
```

```python
@time_me
@debug
def slow(x):
    time.sleep(1)
    return math.sqrt(x)
```

```python
slow(2)
```

## Time me as context manager

```python
class TimeMe:
    def __init__(self):
        ...
    def __enter__(self):
        self.t1 = time.time()

    def __exit__(self, *args):
        self.t2 = time.time()
        print(f'Time used', self.t2-self.t1)
```

```python

def slow(x):
    time.sleep(1)
    return math.sqrt(x)

with TimeMe():
    value= slow(2)

print(value)
```
